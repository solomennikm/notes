```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier
import re

pd.set_option.max_columns = None
pd.set_option.max_rows = None
```

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
train['Upper'] = train['Word'].str.istitle()
test['Upper'] = test['Word'].str.istitle()

train['2end'] = train['Word'].str[-2:].str.lower()
test['2end'] = test['Word'].str[-2:].str.lower()
```

```python
train.head(10)
```

```python
col = [c for c in train if c not in ['Word','Label']]
col2 = [c for c in test if c not in ['Word']]
```

```python
train_full = train[col]
test_full = test[col2]
```

```python
cat_features = np.where(train_full.dtypes == 'object')[0].tolist()
train_full.columns[cat_features]
```

```python
ctb = CatBoostClassifier(random_seed=17, iterations=1000, loss_function='CrossEntropy', eval_metric='AUC')
ctb.fit(train_full, train['Label'], cat_features=cat_features);
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub = sub.reset_index(drop=True)
sub.head()
```

```python
sub['Prediction'] = ctb.predict_proba(test_full)[:,1]
```

```python
sub.head(10)
```

```python
sub.to_csv('submission_easy_catboost.csv', columns=['Id', 'Prediction'], index=False)
```
